```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime, timedelta
```

```python
bilan_actif = pd.read_csv('bilan_actif_psg.csv').drop('Unnamed: 0', axis=1)
bilan_passif = pd.read_csv('bilan_passif_psg.csv').drop('Unnamed: 0', axis=1)
bilan_resultat = pd.read_csv('bilan_compte_resultat_psg.csv').drop('Unnamed: 0', axis=1)
```

```python
bilan_actif['Date de cloture'] = pd.to_datetime(bilan_actif['Date de cloture'], format='%Y-%m-%d')
bilan_passif['Date de cloture'] = pd.to_datetime(bilan_passif['Date de cloture'], format='%Y-%m-%d')
bilan_resultat['Date de cloture'] = pd.to_datetime(bilan_resultat['Date de cloture'], format='%Y-%m-%d')
```

```python
columns_name = {"Date de cloture":"Closing date", "Durée":"Duration", "Devise":"Currency",
                "Capitaux propres":"Shareholders' equity", "Provisions":"Provisions",
                "Dettes":"Debts", "Dettes financières et emprunts":"Financial debts", 
                "Dettes fournisseurs":"Supplier debts", "Dettes fiscales et sociales":"Fiscal debts", 
                "Autres dettes":"Other debts", "Compte de régularisation passif":"Adjustment account",
                "Total passif":"Total liabilities", 
                
                "Actif Immobilisé net":"Long term assets", "immobilisations incorporelles":"Intangible asset",
                "immobilisations corporelles":"Fixed asset", "immobilisations financières":"Investments",
                "Actif circulant net":"Current assets", "stocks et en-cours":"Inventories", 
                "créances":"Account Receivable", "disponibilités : trésorerie":"Liquid Assets", 
                "Total actif":"Total assets"}
```

## Function

```python
from scipy.optimize import curve_fit

def func_deg_3(x, a, b, c, d):
    return  a*x**3 + b*x**2 + c*x + d

def func_deg_2(x, b, c, d):
    return   b*x**2 + c*x + d

def func_deg_4(x, a, b, c, d, e):
    return  a*x**4 + b*x**3 + c*x**2 + d*x + e
```

## Total assets - Long term assets

```python
popt, _ = curve_fit(func_deg_3, np.arange(bilan_actif.shape[0]), bilan_actif["Total assets"].values)
popt
```

```python
plt.plot(bilan_actif.index, func_deg_3(np.arange(bilan_actif.shape[0]), *popt), 'o-', label="Trend Curve")
bilan_actif["Total assets"].plot(marker='o')
plt.legend()
plt.grid()
plt.gca().spines["right"].set_visible(False)
plt.gca().spines["top"].set_visible(False)
plt.savefig('total_asset_both.png')
plt.show()
```

```python
index_ = list(range(2006, 2025))
plt.plot(index_, func_deg_3(np.arange(bilan_actif.shape[0]+5), *popt))
plt.grid()
plt.gca().spines["right"].set_visible(False)
plt.gca().spines["top"].set_visible(False)
plt.xticks([2010, 2015, 2020, 2024],['2010', '2015', '2020', '2024'])
plt.savefig('total_asset_future.png')
plt.show()
```
