```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
start_date = datetime.date(2008, 1, 1)
end_date = datetime.date(2020, 1, 1)
delta = datetime.timedelta(days=364)
index=[]
while start_date <= end_date:
    start_date += delta
    index.append(str(start_date.year)+'-'+str(start_date.year+1))
```

```python
index
```

```python


  
# list of name, degree, score
expense = [10000000,15000000,9000000,107000000, 151000000 ,135000000,49000000, 116000000 ,134000000,238000000 ,227000000,95000000,61000000]
income = [21000000,6000000,7000000,9000000, 4000000,     26000000, 2000000 ,22000000,  59000000  ,98000000  ,114000000 ,105000000 ,5000000   ]
balance_sheet = [income[i]-expense[i] for i in range(len(expense))]
  
# dictionary of lists 
dict = {'expense': expense , 'income': income, 'balance_sheet': balance_sheet} 
    
df = pd.DataFrame(dict,index=index)
    
df 
```

```python
plt.figure(figsize=(15, 8), dpi=80)
plt.plot(df.expense)
plt.plot(df.income)
plt.grid()
plt.legend(df.columns)
plt.title("Evolution of PSG expenses and incomes during Mercatos since 2008-2009")
```

```python
plt.figure(figsize=(15, 8), dpi=80)
plt.plot(df.balance_sheet)
plt.grid()
plt.legend(df.columns[2:],bbox_to_anchor=(0.57, 0.95))
plt.title("Evolution of PSG's balance sheet since 2008-2009")
```
